4. RFM Analysis

```python
current_date = df4["InvoiceDate"].max()+ pd.Timedelta(days = 1)
current_date
```

```python
rfm = df4.groupby("CustomerID").agg({
    'InvoiceDate': lambda x: (current_date - x.max()).days,
    'InvoiceNo': 'count',
    'TotalSales': 'sum'
})
rfm.columns = ['Recency', 'Frequency' , 'Monetary']
rfm.head(5)
```

```python
df4[df4.CustomerID == 12347]['TotalSales'].sum()
```

```python
# Segment Customers based on RFM

rfm['R_segment'] = pd.qcut(rfm['Recency'], 4, labels=[ 4, 3, 2, 1])
rfm['F_segment'] = pd.qcut(rfm['Frequency'], 4, labels=[1, 2, 3, 4])
rfm['M_segment'] = pd.qcut(rfm['Monetary'], 4, labels=[1, 2, 3, 4])
rfm['RFM_score'] = rfm[['R_segment', 'F_segment', 'M_segment']].sum(axis=1)
rfm
```

```python
# Customers with highest RFM score
rfm.sort_values('RFM_score', ascending=False)
```

5. Customer Churn Analysis

```python
df4.head(2)
```

```python
# Create a basket for association rule mining

customer_last_purchase = df4.groupby("CustomerID")["InvoiceDate"].max()
customer_last_purchase.head(5)
```

```python
type(customer_last_purchase)
```

```python
current_date
```

```python
customer_last_purchase = (current_date - customer_last_purchase).dt.days
customer_last_purchase.head(5)
```

```python
# Define Churn thresold (example -  90 days without purchasing)

churn_thresold = 90
churned_customers = customer_last_purchase[customer_last_purchase > churn_thresold]
churned_customers.head(5)
```

```python
# prompt: Distribution Plot: Showing how long ago customers last made a purchase.

import matplotlib.pyplot as plt
import seaborn as sns

# Assuming 'customer_last_purchase' is a pandas Series as calculated in the provided code
# Create the distribution plot
plt.figure(figsize=(10, 6))
sns.distplot(customer_last_purchase, kde=True)  # Use distplot for distribution visualization

plt.title('Distribution of Days Since Last Purchase')
plt.xlabel('Days Since Last Purchase')
plt.ylabel('Density')
plt.show()
```

```python
print("Number of Churned Customers:", len(churned_customers))
plt.figure(figsize=(8, 6))
plt.hist(customer_last_purchase, bins=50, color='green', alpha = 0.7)
plt.axvline(churn_thresold, color='red', linestyle='dashed', linewidth=2)
plt.title("Customer Churn Distribution")
plt.xlabel('Days Since Last Purchase')
plt.ylabel('Number of Customers')
plt.show()
```

```python
import matplotlib.pyplot as plt

# Assuming 'churned_customers' and 'customer_last_purchase' are defined
num_churned = len(churned_customers)
num_stayed = len(customer_last_purchase) - num_churned

# Data
labels = ['Churned', 'Stayed']
values = [num_churned, num_stayed]
colors = ['#FF6F61', '#6B8E23']  # Aesthetic color choices

# Create the bar chart
plt.figure(figsize=(7, 5))
bars = plt.bar(labels, values, color=colors, edgecolor='black', linewidth=1.2)

# Add title and labels with improved styling
plt.title('Customer Churn vs. Stayed', fontsize=14, fontweight='bold', pad=15)
plt.ylabel('Number of Customers', fontsize=12)

# Add value labels above each bar
for bar in bars:
    plt.text(bar.get_x() + bar.get_width()/2, bar.get_height() + 3,
             f'{bar.get_height()}', ha='center', fontsize=12, fontweight='bold', color='black')

# Add gridlines for better readability
plt.grid(axis='y', linestyle='--', alpha=0.6)

# Show the plot
plt.show()
```
